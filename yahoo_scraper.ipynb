```python
import pandas as pd
import numpy as np
import requests
```

```python
pd.set_option('display.max_rows', 2000)
```

```python
def getdata(url):
  r = requests.get(url,headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
  data = pd.read_html(r.text)
  return data
```

```python
stock = "S32"
```

```python
summary_url = f'https://finance.yahoo.com/quote/{stock}.AX?p={stock}.AX&.tsrc=fin-srch'
summary_data = getdata(summary_url)
summary_data
```

```python
data = pd.concat(summary_data)
data.reset_index(drop=True, inplace=True)
data
```

```python
df = data.transpose()
df.columns = df.iloc[0]
df = df.drop(0)
df = df.reset_index(drop=True)
```

```python
historical_url= f'https://finance.yahoo.com/quote/{stock}.AX/history?period1=1518652800&period2=1676419200&interval=1d&filter=history&frequency=1d&includeAdjustedClose=true'

#https://finance.yahoo.com/quote/{stock}/history?p={stock}'
historical_data = getdata(historical_url)
data = historical_data[0]
data = data.iloc[:-1, :]
data = data[data["Open"].str.contains("Dividend") == False]
```

```python
data['Date'] = pd.to_datetime(data['Date'])
numeric_columns = list(data.columns)[1::]
for column_name in numeric_columns:
    data[column_name] = data[column_name].str.replace(',', '')
    data[column_name] = data[column_name].astype(np.float64)
```
